```python
# Write a function that takes the radius of a circle and returns its area.

import math
def get_circle_area(radius):
    return math.pi * (radius ** 2)
print(f"Area of circle (r=5): {get_circle_area(5):.2f}")
```

```python
# Write a function that accepts three numbers and returns the largest one.

def get_largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
print(f"Largest of 10, 25, 7: {get_largest(10, 25, 7)}")
```

```python
# Write a function that takes a string and returns its length without using built-in length functions.

def manual_length(string):
    count = 0
    for char in string:
        count += 1
    return count
print(f"Length of 'Hello': {manual_length('Hello')}")

```

```python
# Write a function that checks whether a number is prime.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(f"Is 17 prime?: {is_prime(17)}")
```

```python
# Write a recursive function to find the nth Fibonacci number.

def fibonacci(n):
    if n <= 0: return 0
    if n == 1: return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
print(f"10th Fibonacci number: {fibonacci(10)}")
```

```python
# Write a recursive function to reverse a string.

def reverse_string(s):
    if len(s) <= 1:
        return s
    return reverse_string(s[1:]) + s[0]
print(f"Reverse of 'Recursion': {reverse_string('Recursion')}")
```

```python
# Write a program to find the sum of all numbers from 1 to n using a for loop.

def sum_to_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
print(f"Sum 1 to 10: {sum_to_n(10)}")
```

```python
# Write a program to count how many even numbers are present in a list.

def count_evens(numbers):
    count = 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count
print(f"Evens in [1, 2, 3, 4, 5, 6]: {count_evens([1, 2, 3, 4, 5, 6])}")
```

```python
# Write a program to print all prime numbers between 1 and 100.

def print_primes_to_100():
    primes = []
    for num in range(1, 101):
        if is_prime(num):
            primes.append(num)
    print("Prime numbers between 1 and 100:")
    print(primes)

print_primes_to_100()
```
